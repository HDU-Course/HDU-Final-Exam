#### 3.编写程序，用户从键盘输入小于1000的整数，对其进行因式分解。例如，10=2×5,60=2×2×3×5

```python
#方法1
n = int(input('请输入一个整数:'))
factor_list = []
remain = n
factor = 2
while remain > 1:
    if remain % factor == 0:
        remain = remain / factor
        factor_list.append(factor)
    else:
        factor += 1
print('%d=%d' % (n,factor_list[0]), end='')
for factor in factor_list[1:]:
    print('×%d' % factor, end='')
print()
```

```python
#方法2（输出修改）
n = int(input('请输入一个整数:'))
factor_list = []
remain = n
factor = 2
while remain > 1:
    if remain % factor == 0:
        remain = remain / factor
        factor_list.append(factor)
    else:
        factor += 1
print('%d=%s' % (n, '×'.join(list(map(str, factor_list)))))
```

#### 4.编写程序，输出所有由1、2、3、4这四个数字组成的素数，并且在每个素数中每个字只使用一次。

```python
import math
def IsPrime(x):
    for i in range(2, int(math.sqrt(x)+1)+1):
        if x % i == 0:
            return False
    return True
```

```python
def PrintIsPrime(x):
    if IsPrime(x):
        print(x)
```

```python
#方法1
alist = [1,2,3,4]
for i in alist:
    pre1 = i
    if IsPrime(pre1):
        print(pre1)
    for j in alist:
        if (i!=j):
            pre2 = pre1 * 10 + j
            if IsPrime(pre2):
                print(pre2)
            for k in alist:
                if (i!=k) and (j!=k):
                    pre3 = pre2 * 10 + k
                    if IsPrime(pre3):
                        print(pre3)
                    for m in alist:
                        if (i!=m) and (j!=m) and (k!=m):
                            pre4 = pre3 * 10 + m
                            if IsPrime(pre4):
                                print(pre4)
```

```python
#方法2
alist = [1,2,3,4]
for i in alist[:]:
    pre1 = i
    PrintIsPrime(pre1)
    alist.remove(i)
    for j in alist[:]:
        pre2 = pre1 * 10 + j
        PrintIsPrime(pre2)
        alist.remove(j)
        for k in alist[:]:
            pre3 = pre2 * 10 + k
            PrintIsPrime(pre3)
            alist.remove(k)
            for m in alist[:]:
                pre4 = pre3 * 10 + m
                PrintIsPrime(pre4)
            alist.append(k)
        alist.append(j)
    alist.append(i)
```

```python
#方法3
alist = [1,2,3,4]
for ii, i in enumerate(alist[:]):
    pre1 = i
    PrintIsPrime(pre1)
    del alist[ii]
    for ji, j in enumerate(alist[:]):
        pre2 = pre1 * 10 + j
        PrintIsPrime(pre2)
        del alist[ji]
        for ki, k in enumerate(alist[:]):
            pre3 = pre2 * 10 + k
            PrintIsPrime(pre3)
            del alist[ki]
            for m in alist[:]:
                pre4 = pre3 * 10 + m
                PrintIsPrime(pre4)
            alist.insert(ki, k)
        alist.insert(ji, j)
    alist.insert(ii, i)
```

#### 6. 查看目录下的star.txt文件，编写程序，用户从键盘输入一个日期，输出该日期对应的星座信息。(提示：读文件方法参照Hw3，文件信息读入一个dict类型的变量）

```python
star_dict = dict()
fp = open('star.txt', 'rt')
for line in fp:
    if '|' in line:
        key,value = line.strip().split('|')
        star_dict[key]=value
fp.close()
star_dict
```

```python
while(True):
    input_date=input('请输入一个日期(回车退出)：')
    if (input_date == ''):
        break
    if input_date in star_dict:
        print(star_dict[input_date])
    else:
        print('找不到该日期的数据，请重新输入')
```
